```python
import sklearn 
import numpy as np 
import pandas as pd 
from datetime import datetime
```

```python
def predict(tup):
    from sklearn.externals import joblib
    import numpy as np
    import pickle
    import math
    #Load random forest model
    rf = joblib.load('random_forests_CE_2.pkl')
    
    #Create query vector
    arr = np.zeros((1,4))[0]
    for i in range(4):
        arr[i] = tup[i]
    arr = np.array(arr)
    
    #return prediction as probability of sepsis in 8 hours
    preds = rf.predict_proba(arr)[:,1]
    return preds[0]



```
